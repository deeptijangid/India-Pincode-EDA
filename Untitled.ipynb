## We will select "Random Regression Forest" as compared to "decision tree" because "Random Regression Forest" is combination of n numbers of "decision tress". Where "decision tree" will give step wise output wheseas, "random regression forest" will take decision in a group which results in increasing in accuracy.
## I would like to select "Random Regression Forest" as Logistic and Random Regression have highest accuracy but my dataset is a complex one so I will not be choosing "Logistic Regression" to train my data
## Also, Choosing "Random Regression Forest" as algorithm to train my data is best as it takes use of multiple decision trees and working is also fast

```python
df.columns
```

```python
x1=df.iloc[:, 0:5]
x2=df.iloc[:, 6:]
X=pd.concat([x1,x2], axis=1)
X = X.drop(['officename', 'circlename', 'regionname', 'divisionname', 'district', 'statename', 'delivery'], axis=1)
X # dropped because these are categorical columns else the model will throw error
# or do label encoding of there columns
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=99)
# Split ratio = 80:20 
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier
#rf=RandomForestClassifier()
# or 
rf=RandomForestClassifier(n_estimators=100, max_depth=None, random_state=42)
#n_estimators(no. of decision tress)=100(default)
#  max_depth(how much in depth analysis)=None(default-till last)(allows the tree to grow fully)
# random_state(fixes randomness of data)=42(more the value less the random, can also change to 99)
rf.fit(X_train, y_train) # trining the algo
rf.score(X_test, y_test) # testing the algo ( accuracy of model)
```

```python
# the accuracy could have been high if we would have done label encoding of all the categorical columns
```

```python
# PRED=rf.predict(X_test) #test
# PRED
# from sklearn.metrics import confusion_matrix
# con=confusion_matrix(y_test,PRED)
# con
```

```python
#Prediction
PRED = rf.predict(X_test)
PRED
#Confusion Matrix
from sklearn.metrics import confusion_matrix, classification_report
con = confusion_matrix(y_test, PRED)
print("Confusion Matrix:\n", con)
# Classification Report
print("\nClassification Report:\n", classification_report(y_test, PRED))
```
